## Generación de Inserts

### Nations

```sql
CREATE TABLE nations (
  id	INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  name	VARCHAR(40)
);

-- migrate:up

INSERT INTO nations (id, name) VALUES (1, 'Albania');
INSERT INTO nations (id, name) VALUES (2, 'Argelia');
...

-- migrate:down

DELETE FROM nations;

```

````python
tmp = ds_players['Nation']
nations = sorted(list(tmp.unique()))

text = '-- migrate:up \n\n'
id = 1
nations_dicc = {}

for n in nations:
  name = n.replace("'", "''")
  text = text + f"INSERT INTO nations (id, name) VALUES ({id}, '{name}');\n"
  nations_dicc[name] = id
  id = id + 1
text = text + '\n-- migrate:down \n\nDELETE FROM nations;'
with open('inserts_nations.sql', 'w') as archivo:
  archivo.write(text)
````

### Leagues

CREATE TABLE leagues (

### Teams

```sql
CREATE TABLE teams (
  id	INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  name	VARCHAR(40),
  league_id	INTEGER NOT NULL,
  FOREIGN KEY (league_id) REFERENCES leagues (id)
);
```

````python
tmp = ds_players[['League', 'Team']]

text = '-- migrate:up \n\n'
id = 1

tmp_lista = []
teams_dicc = {}

for _, row in tmp.iterrows():

  league = row['League'].replace("'", "''")
  team = row['Team'].replace("'", "''")

  if league + team not in tmp_lista:
    tmp_lista.append(league + team)
    league_id = leagues_dicc[league]
    teams_dicc[team] = id
    text = text + f"INSERT INTO teams (id, name, league_id) VALUES ({id}, '{team}', {league_id});\n"
    id = id + 1

text = text + '\n-- migrate:down \n\nDELETE FROM teams;'
with open('inserts_teams.sql', 'w', encoding='utf-8') as archivo:
  archivo.write(text)
````

## Llenar Jugadores

```sql
CREATE TABLE players (
  id	INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  name	VARCHAR(60),
  rank INTEGER,
  weak_foot INTEGER,
  skill_moves INTEGER,
  heigth INTEGER,
  weight INTEGER,
  age INTEGER,
  url VARCHAR(120),
  foot_id INTEGER NOT NULL,
  sex_id INTEGER NOT NULL,
  position_id INTEGER NOT NULL,
  nation_id	INTEGER NOT NULL,
  team_id	INTEGER NOT NULL,
  FOREIGN KEY (foot_id) REFERENCES foots (id),
  FOREIGN KEY (sex_id) REFERENCES sexs (id),
  FOREIGN KEY (position_id) REFERENCES positions (id),
  FOREIGN KEY (nation_id) REFERENCES nations (id)
  FOREIGN KEY (team_id) REFERENCES teams (id)
);

CREATE TABLE common_details (
  id	INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  overall INTEGER, -- OVR
  velocity INTEGER, -- PAC
  shooting INTEGER, -- SHO
  passing INTEGER,  -- PAS
  dribbling INTEGER, -- DRI
  defending INTEGER, -- DEF
  physicality INTEGER, -- PHY
  player_id	INTEGER NOT NULL,
  FOREIGN KEY (player_id) REFERENCES players (id)
);

CREATE TABLE goalkeeper_details (
  id	INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  diving INTEGER,
  handling INTEGER,
  kicking INTEGER,
  positioning INTEGER,
  reflexes INTEGER,
  player_id	INTEGER NOT NULL,
  FOREIGN KEY (player_id) REFERENCES players (id)
);
```

````python
# =====================================
# GENERAR INSERTS:
# - players
# - common_details
# - goalkeeper_details
# =====================================

players_text = '-- migrate:up\n\n'
common_text = '-- migrate:up\n\n'
goalkeeper_text = '-- migrate:up\n\n'

id = 1

foot_dicc = {
    'Right': 1,
    'Left': 2
}

# =====================================
# MALE PLAYERS
# =====================================

tmp = ds_males[
    [
        'Name', 'Rank', 'Weak foot', 'Height', 'Weight', 'Age',
        'Preferred foot', 'Position', 'Nation', 'Team', 'url',
        'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY',
        'GK Diving', 'GK Handling', 'GK Kicking',
        'GK Positioning', 'GK Reflexes'
    ]
]

for _, row in tmp.iterrows():

    name = row['Name'].replace("'", "''")
    rank = row['Rank']
    weak_foot = row['Weak foot']
    height = int(str(row['Height'])[:3])
    weight = int(str(row['Weight']).split('kg')[0])
    age = row['Age']
    url = row['url']

    foot_id = foot_dicc[row['Preferred foot']]
    sex_id = 1

    position = row['Position']
    position_id = positions_dicc[position]

    nation_id = nations_dicc[row['Nation'].replace("'", "''")]
    team_id = teams_dicc[row['Team'].replace("'", "''")]

    # ==========================
    # PLAYERS
    # ==========================

    players_text += (
        f"INSERT INTO players "
        f"(id, name, rank, weak_foot, heigth, weight, age, foot_id, url, "
        f"sex_id, position_id, nation_id, team_id) "
        f"VALUES ("
        f"{id}, "
        f"'{name}', "
        f"{rank}, "
        f"{weak_foot}, "
        f"{height}, "
        f"{weight}, "
        f"{age}, "
        f"{foot_id}, "
        f"'{url}', "
        f"{sex_id}, "
        f"{position_id}, "
        f"{nation_id}, "
        f"{team_id}"
        f");\n"
    )

    # ==========================
    # COMMON DETAILS
    # ==========================

    common_text += (
        f"INSERT INTO common_details "
        f"(overall, velocity, shooting, passing, dribbling, defending, physicality, player_id) "
        f"VALUES ("
        f"{row['OVR']}, "
        f"{row['PAC']}, "
        f"{row['SHO']}, "
        f"{row['PAS']}, "
        f"{row['DRI']}, "
        f"{row['DEF']}, "
        f"{row['PHY']}, "
        f"{id}"
        f");\n"
    )

    # ==========================
    # GOALKEEPER DETAILS
    # ==========================

    if position == 'GK':

        goalkeeper_text += (
            f"INSERT INTO goalkeeper_details "
            f"(diving, handling, kicking, positioning, reflexes, player_id) "
            f"VALUES ("
            f"{row['GK Diving']}, "
            f"{row['GK Handling']}, "
            f"{row['GK Kicking']}, "
            f"{row['GK Positioning']}, "
            f"{row['GK Reflexes']}, "
            f"{id}"
            f");\n"
        )

    id += 1

# =====================================
# FEMALE PLAYERS
# =====================================

tmp = ds_females[
    [
        'Name', 'Rank', 'Weak foot', 'Height', 'Weight', 'Age',
        'Preferred foot', 'Position', 'Nation', 'Team', 'url',
        'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY',
        'GK Diving', 'GK Handling', 'GK Kicking',
        'GK Positioning', 'GK Reflexes'
    ]
]

for _, row in tmp.iterrows():

    name = row['Name'].replace("'", "''")
    rank = row['Rank']
    weak_foot = row['Weak foot']
    height = int(str(row['Height'])[:3])
    weight = int(str(row['Weight']).split('kg')[0])
    age = row['Age']
    url = row['url']

    foot_id = foot_dicc[row['Preferred foot']]
    sex_id = 2

    position = row['Position']
    position_id = positions_dicc[position]

    nation_id = nations_dicc[row['Nation'].replace("'", "''")]
    team_id = teams_dicc[row['Team'].replace("'", "''")]

    # ==========================
    # PLAYERS
    # ==========================

    players_text += (
        f"INSERT INTO players "
        f"(id, name, rank, weak_foot, heigth, weight, age, foot_id, url, "
        f"sex_id, position_id, nation_id, team_id) "
        f"VALUES ("
        f"{id}, "
        f"'{name}', "
        f"{rank}, "
        f"{weak_foot}, "
        f"{height}, "
        f"{weight}, "
        f"{age}, "
        f"{foot_id}, "
        f"'{url}', "
        f"{sex_id}, "
        f"{position_id}, "
        f"{nation_id}, "
        f"{team_id}"
        f");\n"
    )

    # ==========================
    # COMMON DETAILS
    # ==========================

    common_text += (
        f"INSERT INTO common_details "
        f"(overall, velocity, shooting, passing, dribbling, defending, physicality, player_id) "
        f"VALUES ("
        f"{row['OVR']}, "
        f"{row['PAC']}, "
        f"{row['SHO']}, "
        f"{row['PAS']}, "
        f"{row['DRI']}, "
        f"{row['DEF']}, "
        f"{row['PHY']}, "
        f"{id}"
        f");\n"
    )

    # ==========================
    # GOALKEEPER DETAILS
    # ==========================

    if position == 'GK':

        goalkeeper_text += (
            f"INSERT INTO goalkeeper_details "
            f"(diving, handling, kicking, positioning, reflexes, player_id) "
            f"VALUES ("
            f"{row['GK Diving']}, "
            f"{row['GK Handling']}, "
            f"{row['GK Kicking']}, "
            f"{row['GK Positioning']}, "
            f"{row['GK Reflexes']}, "
            f"{id}"
            f");\n"
        )

    id += 1

# =====================================
# MIGRATE DOWN
# =====================================

players_text += '\n-- migrate:down\n\nDELETE FROM players;'
common_text += '\n-- migrate:down\n\nDELETE FROM common_details;'
goalkeeper_text += '\n-- migrate:down\n\nDELETE FROM goalkeeper_details;'

# =====================================
# GENERAR ARCHIVOS
# =====================================

with open('inserts_players.sql', 'w', encoding='utf-8') as archivo:
    archivo.write(players_text)

with open('inserts_common_details.sql', 'w', encoding='utf-8') as archivo:
    archivo.write(common_text)

with open('inserts_goalkeeper_details.sql', 'w', encoding='utf-8') as archivo:
    archivo.write(goalkeeper_text)

print('Archivos generados correctamente.')

````
